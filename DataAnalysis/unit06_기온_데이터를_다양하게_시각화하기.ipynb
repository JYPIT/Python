```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)

aug = []

for row in data:
  month = row[0].split('-')[1]

  if row[-1] != '':
    if month =='08':
      aug.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(aug,bins=10,color='lightblue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
  month = row[0].split('-')[1]

  if row[-1] != '':
    if month =='08':
      aug.append(float(row[-1]))
    if month =='01':
      jan.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(aug,bins=100,color='lightpink',label ='Aug')
plt.hist(jan,bins=100,color='lightblue',label="Jan")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)

aug = []
jan = []
apr = []

for row in data:
  month = row[0].split('-')[1]

  if row[-1] != '':
    if month =='08':
      aug.append(float(row[-1]))
    if month =='01':
      jan.append(float(row[-1]))
    if month =='04':
      apr.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(aug,bins=100,color='lightpink',label ='Aug')
plt.hist(jan,bins=100,color='lightblue',label="Jan")
plt.hist(apr,bins=100,color='lightgreen',label="Apr")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
  day.append([]) # 리스트 내에 리스트 생성

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot') # 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)

plt.figure(figsize=(20,10))
plt.boxplot(day,showfliers=False) #아웃라이어(이상치) 값 생약
plt.show()
```
