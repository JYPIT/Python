###우리 동네 연령별 인구수 result 리스트에 저장하기 -> 시각화를 위해

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv',encoding='cp949')
data = csv.reader(f)

next(data)

result = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:]:
      result.append(int(i))

print(result)
```

###우리 동네 연령별 인구수 데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

plt.figure(dpi = 200) 
plt.style.use('seaborn-white') #ggplot을 주로 사용한다.
plt.plot(result)
plt.show()
```
