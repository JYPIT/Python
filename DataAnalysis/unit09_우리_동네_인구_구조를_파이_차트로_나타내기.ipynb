```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

size = []

name = input("찾고 시은 지역의 이름을 알려주세요 : ")
for row in data:
  if name in row[0]:
    m = 0
    f = 0
    for i in range(101):
      m += int(row[3+i].replace(',',''))
      f += int(row[106+i].replace(',',''))
    break

size.append(m)
size.append(f)

color = ['crimson', 'darkcyan']
label = ['남','여']
plt.style.use('ggplot')
plt.figure(dpi=200)

plt.axis('equal')
plt.pie(size, labels=label, autopct ='%.1f%%',colors=color,startangle=90,explode=(0.1,0))
plt.legend()
plt.title(name+' 지역의 남녀 성별 비율 ')
plt.show()
```
