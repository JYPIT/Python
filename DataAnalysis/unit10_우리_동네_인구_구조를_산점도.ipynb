연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

result = []

name = input("찾고 싶은 지역의 이름을 알려주세요: ")
for row in data :
  if name in row[0] :
    for i in range(3,104) :
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=200)
plt.rc ('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

plt.bar(range(101), result, color= 'r')
plt.title(name+"지역의 연령별 남여 인구 차이")
plt.xlabel("나이")
plt.ylabel("인구 차이")
plt.legend()
plt.show()
```

###제주도의 연령대별 성별 비율을 산점도로 표현하기<br>

남녀 연령별 인구 데이터 scatter() 함수로 표현하기


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
 
%config InlineBackend.figure_format = 'retina'
 
!apt -qq -y install fonts-nanum
 
import matplotlib.font_manager as fm

fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic') 
mpl.font_manager._rebuild()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []
size = []

name = input("찾고 싶은 지역의 이름을 알려주세요 : ")

for row in data :
  if name in row[0]:
    for i in range(3,104):
      m.append(int(row[i].replace(',','')))
      f.append(int(row[i+103].replace(',','')))
    break

plt.figure(dpi=200)
plt.style.use('seaborn')
# alpha 투명도 추가 0~1: 0에 가까울수록 투명, 1에 가까울수록 불투명
plt.scatter(m,f,c=range(101),alpha=0.5,cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g')
plt.show()
```

###원의 크기 -> 인구수의 제곱근

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []
size = []

name = input("찾고 싶은 지역의 이름을 알려주세요 : ")

for row in data :
  if name in row[0]:
    for i in row[3:104]:
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',',''))))
    for i in row[106:]:
      f.append(int(i.replace(',','')))
    break

plt.figure(dpi=200)
plt.style.use('seaborn')
# alpha 투명도 추가 0~1: 0에 가까울수록 투명, 1에 가까울수록 불투명
plt.title(name+" 지역의 성별 인구 그래프")
plt.scatter(m,f,s=size,c=range(101),alpha=0.5,cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g')
plt.xlabel("남성 인구수")
plt.ylabel("여성 인구수")
plt.show()
```
